**Monthly Salary**

```python
df['Monthly Salary'].describe()
```

```python
Q1 = df['Monthly Salary'].quantile(0.25)
Q3= df['Monthly Salary'].quantile(0.75)
IQR= Q3 - Q1
upper = Q3 + IQR*1.5
lower = Q1 - IQR*1.5
```

```python
df[(df['Monthly Salary']>lower)&(df['Monthly Salary']<upper)]
```

the data seems to be in between the quantile range which means the salaries ae good and there are no outliers

```python
sns.histplot(data=df,x='Monthly Salary',palette=sns.color_palette("Paired"))
plt.title('Monthly Salary Distrbution')
plt.xlabel('Monthly Salary')
plt.show()
```

most of the salary is between 1900 to 2250
